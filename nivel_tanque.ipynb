$\frac{dV}{dt}$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Definir los parámetros del sistema para medir el nivel de agua en el tanque
A = np.pi  # Área transversal m^2
g = 9.81   # Gravedad
d = 2      # Diámetro del tanque  (m)
r = d / 2  # Radio del tanque (m)
Q_entrada = 0.5 / 60  # Flujo de entrada en m^3/s

# Definir la función dV_dt
def dV_dt(h, t):
    # Calcular el flujo de salida usando la ecuación de Torricelli
    Q_salida = A * np.sqrt(2 * g * h)
    # La tasa de cambio del volumen es la diferencia entre el flujo de entrada y el flujo de salida
    dV_dt = Q_entrada - Q_salida
    return dV_dt

# Condiciones iniciales
h0 = 3  # Altura inicial del agua en el tanque

# Definir el tiempo de integración
time = np.linspace(0, .4, 1000)  # Tiempo 

# se hace la p intefralara obtener la altura del agua en función del tiempo
altura_liquido = odeint(dV_dt, h0, time)

# Ggraficas
plt.plot(time, altura_liquido)
plt.xlabel('Tiempo (s)')
plt.ylabel('Altura de agua (m)')
plt.title('Altura del agua en en el tanque')
plt.grid(True)
plt.show()

```
